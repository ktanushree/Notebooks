## Metrics: Site Health Score

```python
data = {
    "start_time": "2023-05-03T17:00:00Z",
    "end_time": "2023-05-03T18:00:00Z",
    "interval": "5min",
    "metrics": [
        {"name": "Healthscore","statistics": ["max"], "unit": "gauge"}
    ],
    "filter": {
        "site": ["15803304703800130"]
    }
}
resp = sdk.post.metrics_monitor(data=data)
cloudgenix.jd_detailed(resp)
```
